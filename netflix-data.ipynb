```r
#Creating df for type counts
type_freq <- data.frame("Category" = c("Movie", "TV Show"), "Amount" = c(6131, 2676), "Percents" = c("69.61%", "30.39%"))
```

```r
ggplot(type_freq, aes(x="", y=Amount, fill=Category)) +
    geom_col(color = "black") +
    geom_text(aes(label = Percents),
             position = position_stack(vjust = 0.5)) +
    coord_polar(theta = "y") +
    theme_void()
```

```r
netflix %>%
count(is.na(netflix$country), sort = TRUE)

country_release <- drop_na(netflix, country)
```

```r
country_release %>%
    group_by(country) %>%
    count(country, type) %>%
    arrange(desc(n)) %>%
    head(25)
```

```r
country_release %>%
    count(country, sort = TRUE) %>%
    group_by(country) %>%
    head(25) %>%
    ggplot(aes(x=country, y=n, fill=country)) +
    geom_col(show.legend = FALSE) +
    coord_flip()
    labs(x=NULL, y="Movies and TV Shows Rleased", title="Top 25 Countries on Netflix") +
    theme(axis.text = element_text(angle = 90, hjust = 1, vjust = 0.5)) +
    geom_label(aes(label=n), show.legend=FALSE) +
    scale_y_continuous(expand = c(0, 0), limits = c(0, 3000))
```
